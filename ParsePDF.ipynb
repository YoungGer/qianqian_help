```python
import os
from PyPDF2 import PdfFileReader, PdfFileWriter, PdfFileMerger
import matplotlib.pyplot as plt
%matplotlib inline
```

# patent add signature

```python
base_path = "/Users/yangguangyao/Desktop/patents/signature"

shiyu_path = "/Users/yangguangyao/Desktop/patents/signature/shiyu.pdf"
shiyu = PdfFileReader(shiyu_path)

cnt = 0

for fi in os.listdir(base_path):
    path_i = os.path.join(base_path, fi)
    
    if not path_i[-7:] == "001.pdf":
        continue

    # get output
    output = PdfFileWriter()

    # orig
    org_i = PdfFileReader(path_i)
    for i in range(org_i.getNumPages() - 1):
        output.addPage(org_i.getPage(i))

    # shiyu signature
    output.addPage(shiyu.getPage(cnt))
    
    # save
    with open("/Users/yangguangyao/Desktop/patents/signature/rst/%s" % fi, "wb") as f:
        output.write(f)
        
    cnt += 1
```
