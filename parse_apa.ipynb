# parse to structure DF

```python
def f2(s):
    phase_l = filter(lambda x: len(x)>0, s.strip().split("."))
    phase_l = map(lambda x: x.strip(), phase_l[1:])
    
    author = phase_l[0]
    title = phase_l[1]
    
    postfix = phase_l[2]
    postfix_array  = postfix.split(";")
    
    if len(postfix_array) == 2:
        location_time, page_number = postfix_array
        
        if location_time[-4:-2]=="19" or location_time[-4:-2]=="20":
            year = location_time[-4:]
            location = location_time[:-4]
            try:
                vol, page = page_number.split(":")
            except:
                print page_number
            return [title, author, year, location, vol, page]
        else:
            return None
    else:
        return None
    
def remove_blank_punc(s):
    s1 = s.strip()
    puncs = "!#$%&\'*+,-./:;<=>?@[\\]^_`{|}~"
    
    if s1[0] in puncs:
        s1 = s1[1:]
    if s1[-1] in puncs:
        s1 = s1[:-1]
    return s1
```

```python
df = pd.DataFrame(np.array(map(f2, s_l)))
df.columns = ["title", "author", "year", "loca", "vol", "page"]
```

```python
df.title = df.title.apply(remove_blank_punc)
df.author = df.author.apply(remove_blank_punc)
df.year = df.year.apply(remove_blank_punc)
df.loca = df.loca.apply(remove_blank_punc)
df.vol = df.vol.apply(remove_blank_punc)
df.page = df.page.apply(remove_blank_punc)
```
